#### **CONSULTA**

Consultar historia, uso y proyectos futuros relacionados con Interfaz Cerebro Computador


**Historia**

Las interfaces cerebro–máquina (BCI, Brain-Computer Interface) nacen de la idea de poder controlar dispositivos mediante el pensamiento, algo que primero apareció en la ciencia ficción y luego en la ciencia real.Sus origenes comienzan en el siglo XX, el neurólogo alemán Hans Berger registra por primera vez la actividad eléctrica del cerebro desde allí nace el EEG(electroencefalograma).En ese trabajo, Berger ya identificaba dos tipos de ondas cerebrales: las ondas alfa y beta. Las ondas alfa, con rango de frecuencias entre 8 – 12 Hertz, son más prominentes cuando un sujeto está en estado de relajación con los ojos cerrados y desaparecen cuando se abren los ojos. Las ondas beta, con rango de frecuencias entre 15 – 25 Hertz, están asociadas con un estado de alerta y concentración.

En las décadas de 1950 y 1960, Joe Kamiya y sus colegas de la Universidad de Chicago descubrieron que una persona podía aprender a controlar voluntariamente la intensidad de sus ondas cerebrales, dando origen a la disciplina conocida como neurofeedback.Para 1973 , el término Interfaz Cerebro-Computadora (Brain-Computer Interface – BCI en inglés) fue introducido por primera vez por Jacques J. Vidal, un investigador belga trabajando en la Universidad de California. En el artículo (Vidal, 1973), el autor propuso la idea de utilizar las señales eléctricas del cerebro para permitir a una persona controlar computadoras o prótesis. Sin embargo, no fue hasta los años 1980s que investigadores en Estados Unidos y Europa lograron llevar a la práctica estas ideas. 

En 1988, Farwell y Donchin propusieron e implementaron el primer sistema BCI, denominado “P300 Speller”, que permitía a una persona deletrear texto. En este paradigma BCI se utiliza el pico en la señal EEG que ocurre aproximadamente 300 milisegundos después de la aparición de un estímulo esperado, conocido como potencial relacionado a un evento . En el “P300 Speller”, se presenta al usuario una cuadrícula de 6×6 caracteres en una pantalla, cuyas filas y columnas parpadean al azar una por vez. El usuario debe entonces concentrarse en un carácter específico por vez y contar mentalmente cuando éste parpadea. Estos eventos generan picos ERP en la señal EEG, permitiendo así identificar el carácter por su ubicación en fila y columna.

A principios de los años 90s, Jonathan Wolpaw y colegas en EEUU desarrollaron un sistema BCI que permitía al usuario controlar un cursor en una dimensión. El principio de funcionamiento de este sistema se basaba en neurofeedback, requiriendo que el usuario se entrenara para aprender a controlar la amplitud de sus ondas cerebrales tipo mu (aproximadamente entre 7 – 13 Hertz) en el área sensoriomotora de la corteza cerebral. De esta manera, el sujeto lograba generar ondas con una amplitud baja para mover el cursor hacia abajo (bottom target), o alta para mover el cursor hacia arriba (top target).

A finales de la década de 1990 y principios del siglo XXI, surgieron nuevos paradigmas en las interfaces cerebro-computadora, impulsando un rápido crecimiento en el número de investigadores dedicados al desarrollo de esta tecnología. Un ejemplo notable es el paradigma de BCI basado en potenciales evocados por estímulos visuales, conocidos como SSVEPs (Steady State Visually Evoked Potentials, por sus siglas en inglés). Este enfoque ganó gran popularidad en esa época. En este paradigma, el usuario fija su atención en uno de varios patrones visuales, cada uno de los cuales parpadea a una frecuencia distinta . Así, la señal EEG de la corteza visual se sincroniza con el estímulo externo, permitiendo asociar cada patrón en la señal EEG con un comando deseado. Aunque este paradigma tiene la desventaja de causar fatiga en el usuario, ha sido utilizado con éxito en diversas aplicaciones prácticas hasta la actualidad[1].


**USO**

*1.Comunicación de pacientes con parálisis*

Una interfaz cerebro–computadora (BCI), también conocida como interfaz cerebro–máquina, es una tecnología que permite controlar una computadora u otros dispositivos externos mediante la interpretación de señales cerebrales. Este tipo de sistemas ha sido evaluado principalmente en pacientes con parálisis, como aquellos con lesiones medulares cervicales o con esclerosis lateral amiotrófica. Sin embargo, su aplicación no ha sido estudiada de manera sistemática en la distrofia muscular de Duchenne (DMD), una enfermedad grave caracterizada por la ausencia de distrofina y la consecuente debilidad y atrofia muscular progresiva.

En estudios recientes, se han implementado sistemas BCI basados en el paradigma P300, empleando un método de selección en dos pasos con estímulos visuales tipo parpadeo organizados por regiones. En estos experimentos, los caracteres seleccionados por los usuarios fueron registrados en línea, obteniéndose una precisión promedio del 71,6 % en pacientes con DMD durante el procedimiento de dos pasos[2].

*2. Control de prótesis y dispositivos robóticos*

En este estudio, el BCI se utiliza como un sistema de control para una prótesis de mano, en el cual las señales electroencefalográficas (EEG) del usuario son captadas y procesadas para identificar su intención motora. A través del uso de técnicas de extracción de características y una red neuronal artificial, el sistema clasifica las señales asociadas a movimientos imaginados. Posteriormente, esta información es implementada en una plataforma FPGA, que se encarga de traducir la intención del usuario en movimientos reales de la prótesis, permitiendo así la ejecución de acciones sin necesidad de actividad muscular[3].

*3. Control de dispositivos (computadores, sillas de ruedas, etc.)*

En este estudio, el BCI se presenta como una tecnología que permite establecer una comunicación directa entre el cerebro y dispositivos externos, con aplicaciones principalmente en la comunicación y la rehabilitación de personas con alteraciones motoras. A través del registro de la actividad cerebral, generalmente mediante EEG, el sistema extrae características relevantes de la señal y las traduce en comandos que pueden ser utilizados para interactuar con computadoras u otros dispositivos. De esta manera, el BCI no solo facilita la comunicación en pacientes con movilidad limitada, sino que también contribuye a procesos de rehabilitación neurológica al proporcionar retroalimentación en tiempo real, permitiendo al usuario modular su actividad cerebral y mejorar su desempeño funcional[4].

*4. Rehabilitación de pacientes con accidente cerebrovascular mediante tecnología de interfaz cerebro-computadora (BCI)*

En este estudio, el BCI se emplea como una herramienta de rehabilitación para pacientes que han sufrido un accidente cerebrovascular, especialmente en fases crónicas de la enfermedad. El sistema utiliza señales electroencefalográficas (EEG) para sincronizar la actividad cerebral del paciente con movimientos periféricos, como la estimulación eléctrica funcional de los músculos. Este enfoque permite establecer una conexión entre la intención motora y la ejecución del movimiento, favoreciendo la plasticidad cerebral. Además, el estudio investiga cómo esta interacción activa, en comparación con terapias pasivas, puede generar mayores mejoras en la función motora y promover procesos de reorganización cerebral asociados a la recuperación funcional[5].

**Proyectos Futuros**

*1. BCI portátiles y sensores flexibles (wearables)*

el BCI se proyecta como una tecnología clave para el desarrollo de sistemas portátiles y de uso cotidiano, gracias al avance de sensores electrónicos cerebrales flexibles. Estos dispositivos permiten una adquisición de señales neuronales más precisa, cómoda y no invasiva, lo que abre la posibilidad de trasladar los sistemas BCI desde entornos de laboratorio hacia aplicaciones reales en el ámbito clínico. En este contexto, se plantea que los futuros desarrollos estarán orientados al monitoreo continuo de la actividad cerebral y al tratamiento de enfermedades neurológicas, facilitando una integración más directa entre el usuario y dispositivos externos. Asimismo, el artículo destaca la necesidad de cerrar la brecha entre la investigación experimental y su implementación en la práctica médica, proponiendo el uso de tecnologías más accesibles, adaptables y eficientes para su aplicación en la vida diaria[6].

 *2. Control de robots con BCI no invasiva*

 Hasta el momento, el uso más extendido de las interfaces cerebro–computadora (BCI) ha sido el apoyo a personas con discapacidades motoras graves. Estas tecnologías se aplican, por ejemplo, en pacientes con síndrome de enclaustramiento, permitiéndoles mejorar su capacidad de comunicación, movilidad e independencia. En el campo médico, las BCI hacen posible controlar prótesis como brazos robóticos, utilizar sistemas de escritura asistida o interactuar con distintos dispositivos únicamente a partir de la actividad cerebral. A pesar de estos avances, muchas de estas aplicaciones aún se encuentran en fases experimentales o en desarrollo clínico. En contraste, en áreas como el bienestar y el entretenimiento, ya existen dispositivos comerciales, como auriculares no invasivos, que emplean retroalimentación biológica para ayudar a reducir el estrés.

A futuro, se espera que surjan nuevas aplicaciones de estas tecnologías. Las BCI podrían favorecer el avance de neuroprótesis más sofisticadas, capaces de restaurar funciones sensoriales o facilitar la interacción con sistemas robóticos en tareas complejas. Además, las interfaces bidireccionales podrían permitir formas más avanzadas de comunicación, como el intercambio directo de información entre cerebros o la conexión con entornos digitales. Sin embargo, el desarrollo e implementación de estas tecnologías no solo dependerán del progreso técnico, sino también de aspectos como la aceptación social, la regulación y las implicaciones éticas[7].

*3.BCI aplicada a vehículos autónomos*

La medición de la actividad cerebral puede proporcionar información valiosa que podría aprovecharse también para aumentar las capacidades humanas. Un ejemplo claro de esto es el proyecto Brain-to-Vehicle: Cuando el conductor de un automóvil detecta un obstáculo repentino en la carretera, su actividad cerebral muestra la intención de presionar el freno. Al decodificar estos potenciales de anticipación con una BCI, e identificar así la intención del conductor de frenar, el coche podría empezar la frenada instantáneamente, incluso antes de que se activen los músculos del conductor, lo que podría llegar a ser de gran ayuda para prevenir un accidente.

En 2018, Nissan, en cooperación con Bitbrain, presentó el primer prototipo de interfaz Brain-to-Vehicle, en el que la experiencia de conducción se mejoró al anticipar las decisiones del conductor después de detectar potenciales relacionados con eventos específicos[8].

*4.Chip de grafeno para Parkinson y otras enfermedades*

Inbrain Neuroelectronics busca restaurar la función o movilidad en pacientes con discapacidades. Han diseñado un implante de chip de grafeno que no solo rastrea la actividad cerebral, sino que también la estimula. Además, envía señales mucho más potentes en comparación con los chips metálicos típicos. Fue probado por primera vez en un paciente en 2024 y podría aplicarse a pacientes con enfermedades como el Parkinson.
