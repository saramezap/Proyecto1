**Historia**

Las interfaces cerebro–máquina (BCI, Brain-Computer Interface) nacen de la idea de poder controlar dispositivos mediante el pensamiento, algo que primero apareció en la ciencia ficción y luego en la ciencia real.Sus origenes comienzan en el siglo XX, el neurólogo alemán Hans Berger registra por primera vez la actividad eléctrica del cerebro desde allí nace el EEG(electroencefalograma).En ese trabajo, Berger ya identificaba dos tipos de ondas cerebrales: las ondas alfa y beta. Las ondas alfa, con rango de frecuencias entre 8 – 12 Hertz, son más prominentes cuando un sujeto está en estado de relajación con los ojos cerrados y desaparecen cuando se abren los ojos. Las ondas beta, con rango de frecuencias entre 15 – 25 Hertz, están asociadas con un estado de alerta y concentración.

En las décadas de 1950 y 1960, Joe Kamiya y sus colegas de la Universidad de Chicago descubrieron que una persona podía aprender a controlar voluntariamente la intensidad de sus ondas cerebrales, dando origen a la disciplina conocida como neurofeedback.Para 1973 , el término Interfaz Cerebro-Computadora (Brain-Computer Interface – BCI en inglés) fue introducido por primera vez por Jacques J. Vidal, un investigador belga trabajando en la Universidad de California. En el artículo (Vidal, 1973), el autor propuso la idea de utilizar las señales eléctricas del cerebro para permitir a una persona controlar computadoras o prótesis. Sin embargo, no fue hasta los años 1980s que investigadores en Estados Unidos y Europa lograron llevar a la práctica estas ideas. 

En 1988, Farwell y Donchin propusieron e implementaron el primer sistema BCI, denominado “P300 Speller”, que permitía a una persona deletrear texto. En este paradigma BCI se utiliza el pico en la señal EEG que ocurre aproximadamente 300 milisegundos después de la aparición de un estímulo esperado, conocido como potencial relacionado a un evento . En el “P300 Speller”, se presenta al usuario una cuadrícula de 6×6 caracteres en una pantalla, cuyas filas y columnas parpadean al azar una por vez. El usuario debe entonces concentrarse en un carácter específico por vez y contar mentalmente cuando éste parpadea. Estos eventos generan picos ERP en la señal EEG, permitiendo así identificar el carácter por su ubicación en fila y columna.

A principios de los años 90s, Jonathan Wolpaw y colegas en EEUU desarrollaron un sistema BCI que permitía al usuario controlar un cursor en una dimensión. El principio de funcionamiento de este sistema se basaba en neurofeedback, requiriendo que el usuario se entrenara para aprender a controlar la amplitud de sus ondas cerebrales tipo mu (aproximadamente entre 7 – 13 Hertz) en el área sensoriomotora de la corteza cerebral. De esta manera, el sujeto lograba generar ondas con una amplitud baja para mover el cursor hacia abajo (bottom target), o alta para mover el cursor hacia arriba (top target).

A finales de la década de 1990 y principios del siglo XXI, surgieron nuevos paradigmas en las interfaces cerebro-computadora, impulsando un rápido crecimiento en el número de investigadores dedicados al desarrollo de esta tecnología. Un ejemplo notable es el paradigma de BCI basado en potenciales evocados por estímulos visuales, conocidos como SSVEPs (Steady State Visually Evoked Potentials, por sus siglas en inglés). Este enfoque ganó gran popularidad en esa época. En este paradigma, el usuario fija su atención en uno de varios patrones visuales, cada uno de los cuales parpadea a una frecuencia distinta . Así, la señal EEG de la corteza visual se sincroniza con el estímulo externo, permitiendo asociar cada patrón en la señal EEG con un comando deseado. Aunque este paradigma tiene la desventaja de causar fatiga en el usuario, ha sido utilizado con éxito en diversas aplicaciones prácticas hasta la actualidad[1].


**USO**

*1.Comunicación de pacientes con parálisis*

Una interfaz cerebro–computadora (BCI), también conocida como interfaz cerebro–máquina, es una tecnología que permite controlar una computadora u otros dispositivos externos mediante la interpretación de señales cerebrales. Este tipo de sistemas ha sido evaluado principalmente en pacientes con parálisis, como aquellos con lesiones medulares cervicales o con esclerosis lateral amiotrófica. Sin embargo, su aplicación no ha sido estudiada de manera sistemática en la distrofia muscular de Duchenne (DMD), una enfermedad grave caracterizada por la ausencia de distrofina y la consecuente debilidad y atrofia muscular progresiva.

En estudios recientes, se han implementado sistemas BCI basados en el paradigma P300, empleando un método de selección en dos pasos con estímulos visuales tipo parpadeo organizados por regiones. En estos experimentos, los caracteres seleccionados por los usuarios fueron registrados en línea, obteniéndose una precisión promedio del 71,6 % en pacientes con DMD durante el procedimiento de dos pasos[2].

*2. Control de prótesis y dispositivos robóticos*

En este estudio, el BCI se utiliza como un sistema de control para una prótesis de mano, en el cual las señales electroencefalográficas (EEG) del usuario son captadas y procesadas para identificar su intención motora. A través del uso de técnicas de extracción de características y una red neuronal artificial, el sistema clasifica las señales asociadas a movimientos imaginados. Posteriormente, esta información es implementada en una plataforma FPGA, que se encarga de traducir la intención del usuario en movimientos reales de la prótesis, permitiendo así la ejecución de acciones sin necesidad de actividad muscular[3].

*3. Control de dispositivos (computadores, sillas de ruedas, etc.)*

En este estudio, el BCI se presenta como una tecnología que permite establecer una comunicación directa entre el cerebro y dispositivos externos, con aplicaciones principalmente en la comunicación y la rehabilitación de personas con alteraciones motoras. A través del registro de la actividad cerebral, generalmente mediante EEG, el sistema extrae características relevantes de la señal y las traduce en comandos que pueden ser utilizados para interactuar con computadoras u otros dispositivos. De esta manera, el BCI no solo facilita la comunicación en pacientes con movilidad limitada, sino que también contribuye a procesos de rehabilitación neurológica al proporcionar retroalimentación en tiempo real, permitiendo al usuario modular su actividad cerebral y mejorar su desempeño funcional[4].

*4. Rehabilitación de pacientes con accidente cerebrovascular mediante tecnología de interfaz cerebro-computadora (BCI)*

En este estudio, el BCI se emplea como una herramienta de rehabilitación para pacientes que han sufrido un accidente cerebrovascular, especialmente en fases crónicas de la enfermedad. El sistema utiliza señales electroencefalográficas (EEG) para sincronizar la actividad cerebral del paciente con movimientos periféricos, como la estimulación eléctrica funcional de los músculos. Este enfoque permite establecer una conexión entre la intención motora y la ejecución del movimiento, favoreciendo la plasticidad cerebral. Además, el estudio investiga cómo esta interacción activa, en comparación con terapias pasivas, puede generar mayores mejoras en la función motora y promover procesos de reorganización cerebral asociados a la recuperación funcional[5].

*Consultar sobre actividad cerebral registrada por EEG durante movimientos e imaginación 
de movimiento de las manos y cómo se hace su análisis en sistemas BCI: Electrodos, ritmos 
cerebrales de interés*

La actividad cerebral registrada mediante electroencefalografía (EEG) durante el movimiento y la imaginación de movimiento de las manos se origina principalmente en la corteza motora, donde se generan señales asociadas a la planificación y ejecución del movimiento voluntario. En las interfaces cerebro-computadora (BCI), se emplea comúnmente la imaginación motora (motor imagery, MI), en la cual el usuario imagina mover la mano izquierda o derecha, produciendo patrones de actividad cerebral similares a los del movimiento real[10].

Estas señales EEG se analizan en función de su contenido en frecuencia, clasificándose en bandas como delta (<4 Hz), theta (4–7 Hz), alfa (8–12 Hz), beta (12–30 Hz) y gamma (>30 Hz). En particular, la actividad alfa en la región sensoriomotora se conoce como ritmo mu, y junto con la banda beta constituye el principal foco de estudio en tareas de movimiento e imaginación motora. Diversos estudios han demostrado que durante la imaginación de movimiento de las manos se producen cambios en estas bandas, lo que permite diferenciar entre distintas tareas motoras[10].

Durante estos procesos se observan fenómenos conocidos como desincronización relacionada con el evento (ERD), cuando disminuye la potencia en una banda, y sincronización relacionada con el evento (ERS), cuando aumenta. En tareas de imaginación motora de las manos, típicamente se presenta una disminución de la potencia en la banda mu y cambios en la banda beta, lo cual constituye la base para el análisis e interpretación de la intención del usuario en sistemas BCI[10].

Para registrar esta actividad, se emplean electrodos ubicados sobre el cuero cabelludo siguiendo el sistema internacional 10-20, el cual estandariza las posiciones de medición. En aplicaciones de imaginación motora de las manos, los electrodos más relevantes son C3, C4 y Cz, ya que están situados sobre la corteza sensoriomotora. Estos electrodos permiten captar las variaciones de actividad cerebral asociadas al movimiento de las manos, siendo ampliamente utilizados en estudios de BCI[11].

Además, existe una organización contralateral en el cerebro: la actividad registrada en el electrodo C3 está relacionada principalmente con el movimiento de la mano derecha, mientras que el electrodo C4 se asocia con la mano izquierda. Durante la imaginación motora, se observan cambios opuestos en estas regiones, lo que permite distinguir entre ambas acciones[11].

En los sistemas BCI, el análisis de estas señales se basa en la identificación de cambios en los ritmos mu y beta registrados en estos electrodos, así como en la detección de patrones ERD/ERS, lo que permite clasificar la intención del usuario y traducirla en comandos. De esta manera, la imaginación de movimiento de las manos se convierte en una de las estrategias más utilizadas para el control de dispositivos mediante EEG[11].






*Consultar y discutir los flujos de procesamiento propuestos para estos tipos de proyectos y 
que se puedan realizar en el marco del curso*

Los sistemas de interfaz cerebro-computadora (BCI) basados en EEG siguen un flujo de procesamiento estructurado que permite transformar la actividad cerebral en comandos utilizables. Según la literatura, especialmente en sistemas basados en imaginación motora, este proceso se compone de varias etapas fundamentales[10].

En primer lugar, se realiza la adquisición de la señal EEG, donde se registran las señales cerebrales mediante múltiples electrodos ubicados en el cuero cabelludo. Posteriormente, se lleva a cabo una etapa de preprocesamiento, en la cual se eliminan ruidos y artefactos presentes en la señal, considerando que las señales EEG son altamente no estacionarias y susceptibles a interferencias externas[10].

Luego, se realiza la extracción de características, cuyo objetivo es identificar patrones relevantes dentro de la señal, como cambios en las bandas de frecuencia asociadas a la imaginación motora, especialmente en los ritmos mu y beta. En algunos sistemas, esta etapa se complementa con la selección de características, donde se eligen únicamente aquellas más representativas, con el fin de mejorar la eficiencia y precisión del sistema[10].

Finalmente, se lleva a cabo la clasificación, en la cual se utilizan algoritmos para determinar la intención del usuario a partir de las características extraídas, como por ejemplo distinguir entre la imaginación de movimiento de la mano izquierda o derecha. Este proceso permite traducir la actividad cerebral en comandos que pueden ser utilizados para controlar dispositivos externos[10].

Este flujo de procesamiento constituye la base de los sistemas BCI y puede ser implementado en el marco del curso mediante herramientas de análisis de señales, programación y técnicas de aprendizaje automático, permitiendo el desarrollo de aplicaciones prácticas basadas en la interpretación de señales EEG[10].

Se implementa una función para extraer el número de run desde el nombre del archivo, ya que este identifica la condición experimental de adquisición. Esto permite clasificar correctamente cada señal dentro del experimento.

```python
def obtener_num_run(nombre_archivo):
    """Extrae solo el número de run del nombre del archivo-  Ejemplo esperado: sub-001_task-motor_run-03_eeg.set → devuelve 3"""
    #En este caso RUN 3 seria= Abrir y cerrar puño izquierdo/ puño derecho. (EJECUTANDO LA TAREA 1)
    try:
        run_str = nombre_archivo.split('run-')[1].split('_')[0]
        # Divide el nombre usando 'run-' y toma la parte después
        # Ej: "sub-001_run-03_eeg.set" → ["sub-001_", "03_eeg.set"]
        # Luego corta hasta el siguiente "_" para aislar el número
        # "03_eeg.set" → "03"
        return int(float(run_str))
            # Convierte a número entero (como "03" o "3.0")

    except (IndexError, ValueError):
        # Si el formato no coincide o no se puede convertir,
        # devuelve -1 como indicador de error
        return -1
```

Para permitir una correcta interpretación de las señales EEG dentro del contexto del experimento, se implementa una función de decodificación de eventos. Esta función tiene como objetivo traducir los códigos de eventos presentes en los registros EEG , considerando tanto el tipo de ejecución (real o imaginada) como la condición motora asociada.

La base de datos contiene distintos “runs” experimentales, los cuales definen el tipo de tarea realizada por el sujeto. En este caso, algunos runs corresponden a movimientos reales y otros a imaginería motora, incluyendo tareas como movimiento de mano izquierda, mano derecha, ambos puños o ambos pies.

A partir de esta información, la función asigna una etiqueta semántica a cada evento (T0, T1 y T2), donde:

T0 corresponde a reposo
T1 y T2 representan distintas acciones motoras dependiendo del run, se diferencia entre ejecución real e imaginación motora.

De esta manera, cada segmento de la señal EEG puede ser asociado a una condición experimental específica, lo cual es fundamental para el análisis posterior de la densidad espectral de potencia y la comparación entre estados cerebrales.

En caso de que el evento o el run no cumplan con los patrones esperados, la función retorna la categoría “Desconocido”, garantizando así el manejo de datos inconsistentes sin interrumpir el flujo del procesamiento.

```python
def interpretar_evento(evento, num_run):
    """
    Traduce el código de evento al significado real según el número de run.

    Runs 3,7,11  → Movimiento Real mano izq/der
    Runs 4,8,12  → Imaginería Motora mano izq/der
    Runs 5,9,13  → Movimiento Real ambos puños/pies
    Runs 6,10,14 → Imaginería Motora ambos puños/pies

    T0 = Reposo (siempre)
    T1 = Mano Izquierda o Ambos Puños
    T2 = Mano Derecha o Ambos Pies
    """
    #Definir tipo de acción y significado. 
    if num_run in [3, 7, 11]:
        tipo_accion = 'Real'
        lateralidad = {'T0': 'Reposo', 'T1': 'Mano_Izquierda', 'T2': 'Mano_Derecha'}
    elif num_run in [4, 8, 12]:
        tipo_accion = 'Imaginado'
        lateralidad = {'T0': 'Reposo', 'T1': 'Mano_Izquierda', 'T2': 'Mano_Derecha'}
    elif num_run in [5, 9, 13]:
        tipo_accion = 'Real'
        lateralidad = {'T0': 'Reposo', 'T1': 'Ambos_Punos', 'T2': 'Ambos_Pies'}
    elif num_run in [6, 10, 14]:
        tipo_accion = 'Imaginado'
        lateralidad = {'T0': 'Reposo', 'T1': 'Ambos_Punos', 'T2': 'Ambos_Pies'}
    else:
         # Si el run no está contemplado
        return 'Desconocido'
#Interpretar el evento. 
    for t in ['T0', 'T1', 'T2']:
        # Verifica si el evento termina en T0, T1 o T2
        # Ej: "Stimulus/T1"
        if evento.endswith(t):
            # T0 siempre es reposo
            # Para T1 y T2 combina:
            # lateralidad + tipo de acción
            # Ej: "Mano_Izquierda_Imaginado"
            if t == 'T0':
                return 'Reposo'
            return f'{lateralidad[t]}_{tipo_accion}'
    # Si no coincide con nada
    return 'Desconocido'
```

Para el análisis de las señales EEG, se implementa una función principal encargada de aplicar un flujo completo de preprocesamiento, segmentación y preparación de los datos para su análisis espectral. Este flujo permite transformar señales crudas en representaciones útiles para el estudio de la actividad cerebral asociada a tareas motoras e imaginería motora.

Inicialmente, se realiza la selección de los canales de interés correspondientes a la región sensoriomotora (C3, C4 y Cz), los cuales están directamente relacionados con la actividad cortical involucrada en el control y planificación del movimiento. Esta selección permite focalizar el análisis en las áreas cerebrales más relevantes para la detección de patrones asociados a la ejecución e imaginación motora.

Posteriormente, se realiza una etapa de visualización de la señal original, con el fin de observar su comportamiento antes del procesamiento y establecer una referencia de comparación. A continuación, se aplica una corrección de tendencia (detrend), la cual elimina componentes de baja frecuencia y desplazamientos en la señal que no corresponden a actividad neurofisiológica relevante, evitando así sesgos en el análisis espectral posterior.

En la tercera fase, se aplica un filtrado pasa banda entre 7 y 30 Hz, con el objetivo de aislar los ritmos cerebrales de interés, específicamente las bandas mu (8–12 Hz) y beta (12–30 Hz), las cuales están asociadas a procesos de planificación motora, ejecución de movimiento y fenómenos de sincronización/desincronización cortical (ERD/ERS)[10].

El filtrado se implementa mediante un filtro FIR (Finite Impulse Response) diseñado con el método de ventana (firwin), el cual es ampliamente utilizado en el procesamiento de señales EEG debido a sus propiedades de estabilidad numérica y su capacidad de mantener una respuesta de fase lineal. Estas características son fundamentales en el análisis de señales neurofisiológicas, ya que evitan distorsiones temporales en la actividad cerebral registrada. A diferencia de los filtros IIR (Infinite Impulse Response), los filtros FIR no presentan realimentación, lo que garantiza estabilidad numérica, ausencia de inestabilidades y un comportamiento más controlado en el dominio de la frecuencia [12,13,15].

Adicionalmente, el orden del filtro FIR es determinado automáticamente por la librería MNE-Python en función de la frecuencia de muestreo y los parámetros de diseño del filtro. Este orden controla el compromiso entre selectividad en frecuencia y costo computacional: un mayor orden permite una mejor separación entre bandas, pero incrementa la complejidad del filtrado y el retardo temporal. Por esta razón, su cálculo automático garantiza un balance adecuado entre precisión espectral y eficiencia del procesamiento[13].

Durante esta etapa también se realiza una comparación visual entre la señal original, la señal filtrada y las componentes eliminadas, así como el análisis de la respuesta en frecuencia del filtro mediante su diagrama de Bode. Este análisis permite validar el comportamiento del filtro en el dominio frecuencial, verificando la correcta conservación de la banda de interés (8–30 Hz), la atenuación de componentes fuera de banda y la respuesta de fase aproximadamente lineal, lo cual es fundamental para preservar la integridad temporal de los eventos EEG[13].

Finalmente, se realiza la segmentación de la señal en épocas utilizando los eventos anotados en el registro EEG, considerando únicamente las condiciones relevantes del experimento (T0, T1 y T2). Esta segmentación permite estructurar la señal en ensayos independientes asociados a cada tarea experimental, lo cual es un paso fundamental para el análisis posterior de la densidad espectral de potencia y la comparación entre condiciones cognitivas.

Como salida, la función retorna tanto las épocas segmentadas como las señales intermedias de cada fase del procesamiento, lo que permite evidenciar el efecto de cada etapa sobre la señal original y garantiza la trazabilidad completa del flujo de procesamiento aplicado.

```python
def procesar_eeg_completo(raw, sujeto_id, nombre_archivo, tipo_tarea, mostrar_graficas=True):
#define una función que procesa completamente una señal EEG desde cruda hasta epochs.

    canales_interes = ['C3', 'C4', 'Cz']
    #Define los canales relevantes para actividad motora
    picks = [ch for ch in raw.ch_names if ch.replace('.', '') in canales_interes]
    #Selecciona solo esos canales del EEG (evita puntos como "C3.")

    raw_orig = raw.copy().pick(picks)
    #Copia la señal original SOLO con esos canales (para visualizar)
    raw_res  = raw.copy().pick(picks)
    #Otra copia para procesar (no alterar la original)

    # FASE 1: Señal Original 
    if mostrar_graficas:
        raw_orig.plot(
            title=f"Fase 1: Señal Original – {sujeto_id}",
            duration=15, scalings={'eeg': 50e-6}, show=False
        ) #Grafica la señal original (15 segundos)
        plt.tight_layout()
        plt.show()

    #  FASE 2: Detrend 
    #Extrae los datos EEG como matriz (canales × tiempo)
    data_original = raw_res.get_data().copy()#
    #Quita la media de la señal (elimina desplazamiento DC)
    data_detrend  = mne.filter.detrend(data_original, order=0, axis=1)
    #Reemplaza la señal original  dfinida arriba por la señal sin tendencia
    raw_res._data = data_detrend.copy()

    if mostrar_graficas:
        #Frecuencia de muestreo
        sfreq = raw_res.info['sfreq']
        #Número de puntos para graficar (5 segundos)
        n_pts = int(5 * sfreq)
        #Vector de tiempo en segundos
        tiempo = np.arange(n_pts) / sfreq
        #Selecciona el primer canal
        canal  = raw_res.ch_names[0]

        #Señal original (convertida a microvoltios)
        datos_orig    = data_original[0, :n_pts] * 1e6
        #Señal sin tendencia
        datos_detrend = data_detrend[0, :n_pts]  * 1e6

        #Hace la gráfica de las dos señales. (original y sin tendencia)
        fig, axes = plt.subplots(2, 1, figsize=(14, 6), sharex=True)
        fig.suptitle(f'Fase 2: Detrend – Canal {canal.replace(".", "")} ({sujeto_id})')
        axes[0].plot(tiempo, datos_orig,    color='steelblue')
        axes[0].set_title('Señal Original')
        axes[1].plot(tiempo, datos_detrend, color='darkorange')
        axes[1].set_title('Señal sin tendencia (Detrend)')
        for ax in axes:
            ax.set_ylabel('Amplitud (µV)')
            ax.grid(True, alpha=0.3)
        axes[1].set_xlabel('Tiempo (s)')
        plt.tight_layout()
        plt.show()

    # FASE 3: Filtrado (7–30 Hz) Ya que es la frecuencia que está entre las bandas de interés. 
    raw_res.filter(l_freq=7, h_freq=30, fir_design='firwin', verbose=False)
    #Guarda la señal filtrada.
    data_filtrada = raw_res.get_data().copy()

    if mostrar_graficas:
        sfreq = raw_res.info['sfreq']
        n_pts = int(5 * sfreq)
        tiempo = np.arange(n_pts) / sfreq
        canal  = raw_res.ch_names[0]

        datos_detrend_plot = data_detrend[0, :n_pts] * 1e6
        #Señal sin tendencia 
        datos_filt_plot    = data_filtrada[0, :n_pts] * 1e6
        #Señal filtrada para graficar

        fig, axes = plt.subplots(3, 1, figsize=(14, 8), sharex=True)
        fig.suptitle(f'Fase 3: Filtrado (7–30 Hz) – Canal {canal.replace(".", "")} ({sujeto_id})')
        axes[0].plot(tiempo, datos_detrend_plot, color='darkorange')
        axes[0].set_title('Señal Detrend (entrada al filtro)')
        axes[1].plot(tiempo, datos_filt_plot,    color='green')
        axes[1].set_title('Señal Filtrada (7–30 Hz)')
        axes[2].plot(tiempo, datos_detrend_plot - datos_filt_plot, color='red')
        # Se Muestra lo que eliminó el filtro
        axes[2].set_title('Componentes eliminadas por el filtro')
        for ax in axes:
            ax.set_ylabel('Amplitud (µV)')
            ax.grid(True, alpha=0.3)
        axes[2].set_xlabel('Tiempo (s)')
        plt.tight_layout()
        plt.show()

#FILTRO FIR
        filt_coefs = mne.filter.create_filter(
            data=None, sfreq=sfreq,
            l_freq=7, h_freq=30,
            method='fir', fir_design='firwin'
        ) # Genera los coeficientes del filtro 


        print(f"\nOrden del filtro FIR: {len(filt_coefs) - 1}") #Orden del filtro
        mne.viz.plot_filter(filt_coefs, sfreq, flim=(0.1, 50), fscale='log') #Visualiza el filtro
        plt.show()

        w, h = freqz(filt_coefs, worN=8192, fs=sfreq) #Calcula la respuesta en frecuencia 
        fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)
        fig.suptitle('Diagrama de Bode – Filtro FIR (7–30 Hz)')
        axes[0].plot(w, 20 * np.log10(np.abs(h) + 1e-10))# Magnitud en Decibelios. 
        axes[0].axvspan(7, 30, alpha=0.10, color='blue',  label='Filtro (7–30 Hz)')
        axes[0].axvspan(8, 30, alpha=0.15, color='green', label='Interés (8–30 Hz)')
        axes[0].axhline(-3, linestyle='--', color='gray', label='-3 dB')
        axes[0].set_ylabel('Magnitud (dB)')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        fase = np.unwrap(np.angle(h)) * 180 / np.pi #Fase del filtro
        axes[1].plot(w, fase)
        axes[1].axvspan(7, 30, alpha=0.10, color='blue')
        axes[1].axvspan(8, 30, alpha=0.15, color='green')
        axes[1].set_ylabel('Fase (°)')
        axes[1].set_xlabel('Frecuencia (Hz)')
        axes[1].grid(True, alpha=0.3)
        axes[0].set_xlim(0, 45)
        axes[1].set_xlim(0, 45)
        plt.tight_layout()
        plt.show()

    # --- FASE 4: Segmentación en épocas ---
    events, event_id = mne.events_from_annotations(raw_res, verbose=False) #Obtiene eventos del EEG (T0, T1, T2)

    event_id_filtrado = {
        k: v for k, v in event_id.items()
        if k.endswith(('T0', 'T1', 'T2')) #Filtra solo los eventos relevantes. 
    }

    if len(events) == 0 or len(event_id_filtrado) == 0:
        return None, {}

    epochs = mne.Epochs( #crea segmentos de la señal 
        raw_res, events, event_id=event_id_filtrado,
        tmin=-0.5, tmax=2.0,#tiempo antes y después del evento.
        baseline=(-0.5, 0), #normaliza usando el periodo previo. 
        preload=True, #carga datos en memoria.
        verbose=False 
    )

    if len(epochs) == 0:
        return None, {} #Diccionario con resultados intermedios 

    fases = {
        'original': data_original, #Señal original.
        'detrend':  data_detrend, #Señal sin tendencia. 
        'filtrada': data_filtrada, #Señal filtrada. 
        'ch_names': [ch.replace('.', '') for ch in raw_res.ch_names], #Nombre de canales limpios 
        'sfreq':    raw_res.info['sfreq'] #Frecuencia de muestreo 
    }

    return epochs, fases #Devuelve épocas para la PSD Y las fases para analisis y visualización.
```

*Para cada señal procesada obtener la densidad espectral de potencia en un conjunto 
de canales de interés para las diferentes bandas en las cuales se divide el EEG de 
acuerdo a la revisiones previas*

El análisis cuantitativo de la señal EEG se fundamenta en el cálculo de la densidad espectral de potencia (PSD) mediante la transformada de Fourier utilizando el método de Welch, el cual permite estimar de manera robusta la distribución de energía de la señal en el dominio frecuencial.

De acuerdo con la literatura, este enfoque es ampliamente utilizado en el análisis de EEG debido a su estabilidad frente al ruido y su capacidad para obtener estimaciones consistentes de la potencia espectral en diferentes bandas de frecuencia.

En este trabajo, la señal EEG se divide en bandas de interés neurofisiológico, específicamente mu (8–12 Hz) y beta (12–30 Hz), las cuales han sido ampliamente asociadas con procesos de activación motora, planificación del movimiento y fenómenos de sincronización/desincronización cortical.

Siguiendo la metodología reportada en estudios previos, las medidas de potencia se normalizan dividiendo la potencia de cada banda por la suma de la potencia total dentro del rango de interés (mu + beta). Esta normalización permite obtener una potencia relativa, reduciendo la influencia de variaciones intersujeto y diferencias en amplitud absoluta de la señal[16].

Este procedimiento es consistente con trabajos previos en análisis EEG cuantitativo, donde la normalización espectral se utiliza para mejorar la comparabilidad entre sujetos y condiciones experimentales, así como para enfocar el análisis en la distribución relativa de energía entre bandas funcionalmente relevantes.

```python
def psd_todas_las_bandas(epochs, sujeto_id, tipo_tarea, mostrar_graficas=True):
# Validación de datos: (si no hay epocas o estan vacios, no hace nada)
    if epochs is None or len(epochs) == 0:
        return {}
# Bandas de interes: Mu asociada a reposo motor - Beta asociada a actividad motora. 
    bandas_interes = {
        'Mu':   (8,  13),
        'Beta': (13, 30),
    }
#CÁLCULO DE PSD 
    try:
        #Se usa el método de Welch (promedio de espectros)
        psds_obj = epochs.compute_psd(
            method='welch',
            fmin=7, # fmin=7 para no mostrar el ruido de baja frecuencia
            fmax=30, # solo interesa hasta beta
            n_fft=512, n_per_seg=256, n_overlap=128,
            verbose=False
        )
    except Exception as e:
        print(f"[ERROR] PSD computation failed: {e}")
        return {}
## Si no hay datos, salir
    if psds_obj.get_data().size == 0:
        return {}
#Obtener frecuencias y promedios 
    freqs    = psds_obj.freqs
    #promedia sobre epocas - queda (canales × frecuencias)
    psd_mean = psds_obj.get_data().mean(axis=0)  # (n_canales, n_freqs)

#GRAFICAS PSD  
    if mostrar_graficas:
        colores = {'Mu': '#e15759', 'Beta': '#76b7b2'}
        fig, ax = plt.subplots(figsize=(12, 5))

        #Graficar cada canal 
        for i, ch_name in enumerate(epochs.ch_names):
            #  plot normal, no semilogy, porque los valores son relativos (0 a 1)
            ax.plot(freqs, psd_mean[i],
                    label=ch_name.replace('.', ''), linewidth=1.5)
            
        #Sombrear las bandas MU Y BETA
        for banda_nom, (f_min, f_max) in bandas_interes.items():
            ax.axvspan(f_min, f_max, alpha=0.15,
                       color=colores.get(banda_nom, 'gray'),
                       label=f'{banda_nom} ({f_min}–{f_max} Hz)')
        ax.set_xlabel('Frecuencia (Hz)')
        ax.set_ylabel('Densidad Espectral (µV²/Hz)')  
        ax.set_title(f'PSD – Mu & Beta | {tipo_tarea} ({sujeto_id})')
        ax.legend(fontsize=8, ncol=2)
        ax.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

#Extraccion de FEAURES 
    resultado = {}
    for i, ch_name in enumerate(epochs.ch_names):
        ch = ch_name.replace('.', '')

        #Total = solo Mu + Beta (8–30 Hz) (Entonces es solo de 8 a 30 Hz por que ya se filtró)
        potencia_total = np.sum(psd_mean[i])  # ya que fmin=8 y fmax=30

        for banda_nom, (f_min, f_max) in bandas_interes.items():
            #indices de frecuencia dentro de la banda.
            idx_banda = np.logical_and(freqs >= f_min, freqs <= f_max)
            #potencia de esa banda.
            potencia_banda = np.sum(psd_mean[i, idx_banda])
            # Potencia relativa 
            resultado[f'{banda_nom}_{ch}'] = potencia_banda / potencia_total

    return resultado
```

*Crear una rutina que aplique sobre todos los archivos de la base de datos las rutinas

nombre sujeto, la condición evaluada y la densidad espectral de potencia obtenida*

Para el procesamiento global de la base de datos EEG, se implementa una rutina automatizada encargada de recorrer todos los sujetos y registros disponibles, con el fin de aplicar de manera consistente el flujo de preprocesamiento definido previamente.

Esta función permite integrar en un único procedimiento las etapas de carga de datos, selección de canales, preprocesamiento, segmentación en épocas y cálculo de la densidad espectral de potencia (PSD), garantizando la homogeneidad en el tratamiento de todas las señales.

Cada archivo EEG es procesado individualmente, y posteriormente segmentado en eventos correspondientes a las condiciones experimentales (T0, T1 y T2). A cada segmento se le aplica el cálculo de la PSD mediante el método de Welch, obteniendo la potencia en las bandas de interés (mu y beta) para los canales seleccionados.

Los resultados obtenidos se almacenan en un dataframe estructurado, en el cual cada fila representa una observación correspondiente a un sujeto, una condición experimental y un evento específico. Las columnas incluyen la identificación del sujeto, el número de run, la condición evaluada y las potencias espectrales por banda y canal.

Esta organización permite realizar análisis comparativos entre sujetos y condiciones, facilitando el estudio de diferencias en la actividad cerebral asociada a tareas motoras e imaginería motora, y asegurando la trazabilidad completa del procesamiento aplicado a cada señal.

```python
def construir_dataframe_completo(base_path, incluir_baseline=False):
# Función principal que genera el dataset completo
#base_path: ruta donde están los sujetos
#incluir_baseline: decide si usar runs 1 y 2 (reposo puro)

    filas = []
#Lista donde se guardará cada fila del DataFrame (cada ejemplo)

    if not os.path.exists(base_path):
        #verifica que la carpeta exista 
        print(f"[ERROR] No se encontró la carpeta: {base_path}")
        #si no existe devuelve Dataframe vacio 
        return pd.DataFrame()

    sujetos = sorted([s for s in os.listdir(base_path)
                      if os.path.isdir(os.path.join(base_path, s))])
#Lista todos los sujetos (carpetas)

    primer_sujeto = sujetos[0] if sujetos else None
    primer_archivo_mostrado = False
#Variables para mostrar gráficas solo una vez 

    for sujeto in sujetos:
        #Recorre los 109 sujetos
        ruta_sujeto = os.path.join(base_path, sujeto)
        #Ruta completa del sujeto
        for archivo in sorted(os.listdir(ruta_sujeto)):
            #Recorre archivos del sujeto
            if not archivo.endswith('.set'):
                # Solo usa archivos EEG (.set)
                continue

            num_run = obtener_num_run(archivo)
            #Extrae número de run
            if not incluir_baseline and num_run in [1, 2]:
                continue
                #Opcional: ignora baseline

            mostrar = (sujeto == primer_sujeto and not primer_archivo_mostrado)
            #Solo muestra gráficas para el primer archivo

            try:
                ruta_completa = os.path.join(ruta_sujeto, archivo) #Carga el EEG con MNE-Python
                raw = mne.io.read_raw_eeglab(ruta_completa, preload=True, verbose=False)

                with warnings.catch_warnings(): 
                    warnings.simplefilter("ignore")
                    epochs, fases = procesar_eeg_completo(
                        raw, sujeto, archivo, '',
                        mostrar_graficas=mostrar
                    ) #Ejecuta pipeline: selección de canales, detrend, filtrado, segmentación

                if epochs is None:
                    print(f"[INFO] Sin epochs válidos en {sujeto}/{archivo}")
                    continue
                    #Si no hay datos válidos , saltar
                if mostrar:
                    primer_archivo_mostrado = True
                    #Evita mostrar más gráficas

                # Una fila por cada evento (T0, T1, T2)
                for evento in epochs.event_id.keys():
                    #Recorre T0, T1 Y T2. 
                    epochs_evento = epochs[evento]
                    #Obtiene solo esos segmentos.

                    if len(epochs_evento) == 0:
                        continue
                    #Si no hay datos, saltar
                    tarea_interpretada = interpretar_evento(evento, num_run)
                    # Convierte T1 MANO IZQUIERDA, T2 MANO DERECHA , Según el RUN.
                    potencias = psd_todas_las_bandas(
                        epochs_evento, sujeto, evento,
                        mostrar_graficas=mostrar
                    ) #Calcula potencias en mu y beta. (Basado en la densidad espectral de potencia)

                    if not potencias:
                        continue
                    # Si falla, saltar.

                    filas.append({
                        'Registro':   sujeto,
                        'Run':        num_run,
                        'Evento_Raw': evento,
                        'Tarea':      tarea_interpretada,
                        **potencias
                    }) #Cada fila contiene SUJETO, RUN, EVENTO ORIGINAL, TAREA INTERPRETADA, FEATURES (MU Y BETA POR CANAL)

            except Exception as e:
                print(f"[SKIP] {sujeto}/{archivo}: {type(e).__name__} – {e}")
                #Manejo de errores, si falla rompe todo. 
                continue
                #Para seguir con el siguiente archivo y hacer esto para todos los sujetos. 

    df = pd.DataFrame(filas) #Convierte lista en Dataframe
    if df.empty:
        print("[INFO] DataFrame vacío, no se generaron resultados.") #Si no hay datos, avisar.
        return df

    cols_fijas  = ['Registro', 'Run', 'Evento_Raw', 'Tarea'] #Columnas principales.
    cols_bandas = sorted([c for c in df.columns if c not in cols_fijas]) #Columnas de features MU Y BETA. 
    return df[cols_fijas + cols_bandas] #Evuelve Dataframe ordenado.
```

## **Análisis de resultados**

### *Hipótesis 1* : Supresión del ritmo Mu
H₀: La densidad de potencia relativa en las bandas Mu (Mu_C3, Mu_C4, Mu_Cz) es igual entre las condiciones de Reposo, Imaginado y Real.
(μ_Reposo = μ_Imaginado = μ_Real)

H₁: La densidad de potencia relativa Mu es significativamente menor durante tareas motoras reales que durante tareas imaginadas y reposo.
(μ_Real < μ_Imaginado ≤ μ_Reposo)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import warnings
warnings.filterwarnings("ignore")

ORDEN     = ["Reposo", "Imaginado", "Real"]
MU_COLS   = ["Mu_C3", "Mu_C4", "Mu_Cz"]
ALPHA     = 0.05          # Nivel de significancia

# 1.  CARGA Y PREPARACIÓN DE DATOS

def cargar_datos(ruta_csv: str = "resultados_completos.csv") -> pd.DataFrame:
    """
    Carga el CSV generado por construir_dataframe_completo() y añade
    la columna 'Tipo' que clasifica cada fila como:
        'Real'     tarea motora con movimiento físico
        'Imaginado' tarea de imaginería motora
        'Reposo'    línea base sin movimiento

    Parámetros
    ruta_csv : str
        Ruta al archivo CSV exportado por el pipeline EEG.

    Retorna
    pd.DataFrame con columna 'Tipo' añadida.
    """
    df = pd.read_csv(ruta_csv)

    def clasificar_tipo(tarea: str) -> str:
        if "Real"     in str(tarea): return "Real"
        if "Imaginado" in str(tarea): return "Imaginado"
        return "Reposo"

    df["Tipo"] = df["Tarea"].apply(clasificar_tipo)
    return df

# 2.  ANÁLISIS ESTADÍSTICO

def calcular_eta_cuadrado(F: float, df_between: int, df_within: int) -> float:
    """
    Calcula eta² parcial como medida del tamaño de efecto del ANOVA.
    Interpretación:  η² ≈ 0.01 pequeño | 0.06 mediano | 0.14 grande

    Parámetros
    ----------
    F          : estadístico F del ANOVA
    df_between : grados de libertad entre grupos (k-1)
    df_within  : grados de libertad dentro de grupos (N-k)
    """
    ss_between = F * df_between
    ss_total   = ss_between + df_within
    return ss_between / ss_total


def analizar_hipotesis_1(df: pd.DataFrame) -> dict:
    """
    Ejecuta el análisis estadístico completo para la Hipótesis 1:

    Por cada canal Mu (Mu_C3, Mu_C4, Mu_Cz):
        1. ANOVA de un factor (Tipo de tarea: Reposo / Imaginado / Real)
        2. Post-hoc de Tukey si el ANOVA es significativo
        3. Tamaño de efecto eta²

    Parámetros
    ----------
    df : DataFrame con columnas Mu_C3, Mu_C4, Mu_Cz y 'Tipo'.

    Retorna
    -------
    dict con resultados por canal.
    """
    resultados = {}


    for canal in MU_COLS:
        # Extraer grupos
        grupos_data = {g: df[df["Tipo"] == g][canal].dropna() for g in ORDEN}
        n_total     = sum(len(v) for v in grupos_data.values())
        k           = len(ORDEN)

        # ANOVA de un factor 
        F, p_anova = stats.f_oneway(*grupos_data.values())
        eta2       = calcular_eta_cuadrado(F, df_between=k-1, df_within=n_total-k)

        sig_label  = "Significativo" if p_anova < ALPHA else "No significativo"
        print(f"\n  Canal: {canal}")
        print(f"    ANOVA → F = {F:.3f} | p = {p_anova:.4f} | η² = {eta2:.3f}  {sig_label}")

        medias = {g: v.mean() for g, v in grupos_data.items()}
        for g, m in medias.items():
            print(f"      Media {g:>10s}: {m:.5f}")

        # ── Post-hoc Tukey (solo si ANOVA es significativo) ───────────
        tukey_result = None
        if p_anova < ALPHA:
            valores  = df[df["Tipo"].isin(ORDEN)][canal].dropna()
            etiquetas= df.loc[valores.index, "Tipo"]
            tukey_result = pairwise_tukeyhsd(valores, etiquetas, alpha=ALPHA)
            print(f"\n    Post-hoc Tukey (α={ALPHA}):")
            print("   ", str(tukey_result.summary()).replace("\n", "\n    "))

        resultados[canal] = {
            "F": F, "p_anova": p_anova, "eta2": eta2,
            "medias": medias, "grupos_data": grupos_data,
            "tukey": tukey_result, "significativo": p_anova < ALPHA
        }

    return resultados

# 3.  VISUALIZACIONES

def plot_boxplot(df: pd.DataFrame, resultados: dict):
    """
    Genera un boxplot con jitter para cada canal Mu, comparando los tres
    tipos de tarea. Anota el p-valor del ANOVA en cada panel.
    """
    fig, axes = plt.subplots(1, 3, figsize=(15, 6), sharey=False)
    fig.suptitle(
        "H1 — Densidad de Potencia Relativa Mu por Tipo de Tarea\n"
        "(Boxplot + observaciones individuales)",
        fontsize=13, fontweight="bold"
    )

    for ax, canal in zip(axes, MU_COLS):
        res = resultados[canal]

        # Boxplot base
        sns.boxplot(
            data=df, x="Tipo", y=canal, order=ORDEN,
            palette="Set2", width=0.5, linewidth=1.4,
            flierprops=dict(marker="", alpha=0),   # ocultamos outliers del box
            ax=ax
        )
        # Puntos individuales (jitter) para ver distribución real
        sns.stripplot(
            data=df, x="Tipo", y=canal, order=ORDEN,
            palette="Set2", alpha=0.35, size=3.5, jitter=True, ax=ax
        )

        # Anotar p-valor y η²
        # Anotar p-valor y η²
        p  = res["p_anova"]
        e2 = res["eta2"]
        ax.set_title(
        f"{canal}\np = {p:.4f}  |  η² = {e2:.3f}",
        fontsize=11
        )
        ax.set_xlabel("Tipo de tarea")
        ax.set_ylabel("Densidad de potencia relativa")

    plt.tight_layout()
    plt.show()

# 4.  PUNTO DE ENTRADA

if __name__ == "__main__":

    # 1. Cargar datos
    df = cargar_datos("resultados_completos.csv")
    print(f"  Dataset cargado: {df.shape[0]} filas × {df.shape[1]} columnas")
    print(f"  Tipos de tarea encontrados: {df['Tipo'].value_counts().to_dict()}")

    # 2. Análisis estadístico
    resultados = analizar_hipotesis_1(df)

    # 3. Visualización
    plot_boxplot(df, resultados)
```

### *Hipótesis 2* : Lateralización hemisférica

H₀: El índice de lateralización (C4 − C3) / (C4 + C3) no difiere significativamente entre tareas de mano derecha e izquierda, es decir, no hay asimetría hemisférica dependiente de la mano usada.
(LI_Derecha = LI_Izquierda)

H₁: Las tareas de mano derecha producen un LI negativo (dominio C3, hemisferio izquierdo) y las de mano izquierda un LI positivo (dominio C4, hemisferio derecho), siendo esta diferencia significativa.
(LI_Derecha < 0 < LI_Izquierda)

```python

#  Configuración global 
ALPHA        = 0.05
BANDAS       = ["Beta", "Mu"]       # Se calculará LI para cada banda
sns.set_theme(style="whitegrid", font_scale=1.1)


# 1.  CARGA Y PREPARACIÓN DE DATOS

def cargar_datos(ruta_csv: str = "resultados_completos.csv") -> pd.DataFrame:
    """
    Carga el CSV del pipeline EEG y añade columnas auxiliares:
      - 'Tipo'  : Real / Imaginado / Reposo
      - 'Mano'  : Derecha / Izquierda / Ambos / Reposo
      - 'LI_Beta', 'LI_Mu' : Índices de Lateralización por banda

    Parámetros
    ----------
    ruta_csv : str — ruta al archivo resultados_completos.csv

    Retorna
    -------
    pd.DataFrame enriquecido.
    """
    df = pd.read_csv(ruta_csv)

    # ── Clasificar tipo de tarea ──────────────────────────────────────
    def clasificar_tipo(tarea: str) -> str:
        if "Real"      in str(tarea): return "Real"
        if "Imaginado" in str(tarea): return "Imaginado"
        return "Reposo"

    # ── Clasificar mano ───────────────────────────────────────────────
    def clasificar_mano(tarea: str) -> str:
        t = str(tarea)
        if "Derecha"   in t: return "Derecha"
        if "Izquierda" in t: return "Izquierda"
        if "Ambos"     in t: return "Ambos"
        return "Reposo"

    df["Tipo"] = df["Tarea"].apply(clasificar_tipo)
    df["Mano"] = df["Tarea"].apply(clasificar_mano)

    # ── Calcular Índice de Lateralización para Beta y Mu ─────────────
    #   Fórmula: LI = (ipsilateral − contralateral) / (ipsilateral + contralateral)
    #   Usamos C4 como hemisferio derecho y C3 como hemisferio izquierdo.
    #   LI > 0 → mayor potencia en C4 (dominio derecho)
    #   LI < 0 → mayor potencia en C3 (dominio izquierdo)
    for banda in BANDAS:
        c3_col = f"{banda}_C3"
        c4_col = f"{banda}_C4"
        denom  = df[c4_col] + df[c3_col]
        df[f"LI_{banda}"] = np.where(
            denom != 0,
            (df[c4_col] - df[c3_col]) / denom,
            np.nan
        )

    return df

# 2.  ANÁLISIS ESTADÍSTICO
def cohen_d(grupo_a: pd.Series, grupo_b: pd.Series) -> float:
    """
    Calcula la d de Cohen entre dos grupos independientes.
    Interpretación: |d| < 0.2 trivial | 0.5 mediano | 0.8 grande

    Parámetros
    ----------
    grupo_a, grupo_b : Series con las observaciones de cada grupo.
    """
    n_a, n_b   = len(grupo_a), len(grupo_b)
    s_pooled   = np.sqrt(
        ((n_a - 1) * grupo_a.std(ddof=1)**2 + (n_b - 1) * grupo_b.std(ddof=1)**2)
        / (n_a + n_b - 2)
    )
    return (grupo_a.mean() - grupo_b.mean()) / s_pooled if s_pooled != 0 else np.nan


def analizar_hipotesis_2(df: pd.DataFrame) -> dict:
    """
    Prueba estadística de la Hipótesis 2 para cada banda (Beta, Mu):

        1. Filtra filas de Mano Derecha e Izquierda (excluye Ambos y Reposo)
        2. t-test independiente sobre el LI entre los dos grupos de mano
        3. Calcula Cohen's d
        4. Imprime tabla de resultados

    También verifica la hipótesis de dirección:
        - Mano Derecha → esperamos LI < 0  (dominio C3)
        - Mano Izquierda → esperamos LI > 0 (dominio C4)

    Parámetros
    df : DataFrame con columnas LI_Beta, LI_Mu, Mano.

    Retorna
    diccionario con resultados por banda y por tipo de tarea (Real / Imaginado).
    """
    resultados = {}


    # Analizar tanto el conjunto completo como separando Real e Imaginado
    subsets = {
        "Todas las tareas": df,
        "Solo Real":        df[df["Tipo"] == "Real"],
        "Solo Imaginado":   df[df["Tipo"] == "Imaginado"],
    }

    for subset_nombre, subset_df in subsets.items():
        print(f"\n  ── Subconjunto: {subset_nombre} ──")
        resultados[subset_nombre] = {}

        for banda in BANDAS:
            li_col    = f"LI_{banda}"
            derecha   = subset_df[subset_df["Mano"] == "Derecha"][li_col].dropna()
            izquierda = subset_df[subset_df["Mano"] == "Izquierda"][li_col].dropna()

            # Necesitamos al menos 2 observaciones por grupo
            if len(derecha) < 2 or len(izquierda) < 2:
                print(f"    {banda}: datos insuficientes para comparar.")
                continue

            t_stat, p_val = stats.ttest_ind(derecha, izquierda)
            d             = cohen_d(derecha, izquierda)
            sig           = "SI" if p_val < ALPHA else "NO"

            # Verificar dirección esperada
            dir_derecha   = "✓ (LI<0, domina C3)" if derecha.mean()   < 0 else "✗ (LI>0, domina C4)"
            dir_izquierda = "✓ (LI>0, domina C4)" if izquierda.mean() > 0 else "✗ (LI<0, domina C3)"

            print(f"\n    Banda {banda}:")
            print(f"      t = {t_stat:+.3f} | p = {p_val:.4f} | d = {d:+.3f}  {sig}")
            print(f"      Media LI Mano Derecha   = {derecha.mean():+.5f}  {dir_derecha}")
            print(f"      Media LI Mano Izquierda = {izquierda.mean():+.5f}  {dir_izquierda}")
            print(f"      N Derecha={len(derecha)}, N Izquierda={len(izquierda)}")

            resultados[subset_nombre][banda] = {
                "t": t_stat, "p": p_val, "d": d,
                "significativo": p_val < ALPHA,
                "derecha": derecha, "izquierda": izquierda
            }

    return resultados

# 3.  VISUALIZACIONES
def plot_li_boxplot(df: pd.DataFrame, resultados: dict):
    """
    Boxplot del Índice de Lateralización (LI) para Beta y Mu,
    separado por mano (Derecha / Izquierda) y tipo de tarea
    (Real / Imaginado). Incluye línea de referencia en LI=0.
    """
    sub = df[df["Mano"].isin(["Derecha", "Izquierda"])].copy()

    fig, axes = plt.subplots(1, 2, figsize=(13, 6), sharey=False)
    fig.suptitle(
        "H2 — Índice de Lateralización (LI) por Mano y Tipo de Tarea\n"
        "LI < 0 → domina C3 (hemis. izq.)  |  LI > 0 → domina C4 (hemis. der.)",
        fontsize=12, fontweight="bold"
    )

    for ax, banda in zip(axes, BANDAS):
        li_col = f"LI_{banda}"
        res_todas = resultados.get("Todas las tareas", {}).get(banda, {})

        sns.boxplot(
            data=sub, x="Mano", y=li_col, hue="Tipo",
            hue_order=["Real", "Imaginado"],
            palette={"Real": "green" , "Imaginado": "orange" },
            order=["Derecha", "Izquierda"],
            width=0.55, linewidth=1.4,
            flierprops=dict(marker="o", markersize=3, alpha=0.4),
            ax=ax
        )
        # Línea de referencia LI = 0
        ax.axhline(0, color="black", linestyle="--", linewidth=1.2, alpha=0.7,
                   label="LI = 0 (sin lateralización)")

        # Anotar p-valor global
        if res_todas:
            p_txt = f"p = {res_todas['p']:.4f}"
            color_p = "green" if res_todas["significativo"] else "red"
            ax.text(0.97, 0.97, p_txt, transform=ax.transAxes,
                    ha="right", va="top", fontsize=10, color=color_p,
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec=color_p, alpha=0.8))

        ax.set_title(f"Banda {banda}", fontsize=11, fontweight="bold")
        ax.set_xlabel("Mano activa")
        ax.set_ylabel(f"LI_{banda}")
        ax.legend(title="Tipo", loc="lower right", fontsize=9)

    plt.tight_layout()
    plt.savefig("H2_LI_boxplot.png", dpi=150, bbox_inches="tight")
    plt.show()



def plot_topomapa_simplificado(df: pd.DataFrame):
    """
    Representación esquemática del cuero cabelludo (vista cenital) con
    los tres electrodos de interés (C3, Cz, C4) coloreados según la
    diferencia de potencia media entre Mano Derecha e Izquierda.

    NOTA: es un topomapa cualitativo simplificado, no un EEG topoplot
    de librería MNE.
    """
    sub      = df[df["Mano"].isin(["Derecha", "Izquierda"])]
    electrodos = {
        "C3":  (-0.35, 0.0),   # izquierdo
        "Cz":  ( 0.0,  0.0),   # central
        "C4":  ( 0.35, 0.0),   # derecho
    }

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    fig.suptitle(
        "H2 — Topomapa esquemático: Diferencia de potencia Mu\n"
        "(Mano Derecha − Mano Izquierda por electrodo)",
        fontsize=12, fontweight="bold"
    )

    for ax, banda in zip(axes, BANDAS):
        difs = {}
        for elec in ["C3", "Cz", "C4"]:
            col = f"{banda}_{elec}"
            med_der = sub[sub["Mano"] == "Derecha"][col].mean()
            med_izq = sub[sub["Mano"] == "Izquierda"][col].mean()
            difs[elec] = med_der - med_izq   # positivo = más en Derecha

        vmax = max(abs(v) for v in difs.values()) + 1e-6
        cmap = plt.cm.RdBu_r

        # Dibujar para el cráneo
        circulo = plt.Circle((0, 0), 0.55, color="#e8d5b7", lw=2.5,
                              ec="#8B7355", fill=True, zorder=1)
        ax.add_patch(circulo)
        # Nariz
        ax.annotate("", xy=(0, 0.58), xytext=(0, 0.53),
                    arrowprops=dict(arrowstyle="-|>", color="#8B7355", lw=1.5))

        # Dibujar electrodos
        for elec, (x, y) in electrodos.items():
            val   = difs[elec]
            color = cmap((val + vmax) / (2 * vmax))
            circ  = plt.Circle((x, y), 0.08, color=color, ec="black", lw=1.5, zorder=3)
            ax.add_patch(circ)
            ax.text(x, y + 0.0, elec, ha="center", va="center",
                    fontsize=10, fontweight="bold", color="black", zorder=4)
            ax.text(x, y - 0.12, f"{val:+.4f}", ha="center", va="center",
                    fontsize=8, color="black", zorder=4)

        # Colorbar
        sm = plt.cm.ScalarMappable(cmap=cmap,
                                   norm=plt.Normalize(vmin=-vmax, vmax=vmax))
        sm.set_array([])
        plt.colorbar(sm, ax=ax, fraction=0.04, pad=0.04,
                     label="Δ Potencia (Derecha − Izquierda)")

        ax.set_xlim(-0.75, 0.75)
        ax.set_ylim(-0.75, 0.75)
        ax.set_aspect("equal")
        ax.axis("off")
        ax.set_title(f"Banda {banda}", fontsize=11, fontweight="bold")

    plt.tight_layout()
    plt.savefig("H2_topomapa.png", dpi=150, bbox_inches="tight")
    plt.show()


# 4.  PUNTO DE ENTRADA
if __name__ == "__main__":

    # 1. Cargar y preparar datos
    df = cargar_datos("resultados_completos.csv")
    print(f"  Dataset cargado: {df.shape[0]} filas × {df.shape[1]} columnas")
    dist_mano = df["Mano"].value_counts().to_dict()
    print(f"  Distribución por mano: {dist_mano}")

    # 2. Análisis estadístico
    resultados = analizar_hipotesis_2(df)

    # 3. Visualizaciones
    plot_li_boxplot(df, resultados)
    plot_topomapa_simplificado(df)
```

### *Hipótesis 3:* Variabilidad inter-sujeto

H₀: No existen diferencias significativas entre sujetos (sub_1–sub_9) en la magnitud de supresión de Mu durante imaginería motora.
(μ_S1 = μ_S2 = ... = μ_S9)

H₁: Al menos un sujeto presenta una magnitud de supresión de Mu durante imaginería motora significativamente diferente a los demás, reflejando variabilidad en la capacidad individual de imaginación motora.
(Al menos una μᵢ ≠ μⱼ)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import warnings
warnings.filterwarnings("ignore")

# ─── Configuración global ─────────────────────────────────────────────────────
MU_COLS   = ["Mu_C3", "Mu_C4", "Mu_Cz"]
BETA_COLS = ["Beta_C3", "Beta_C4", "Beta_Cz"]
ALL_COLS  = MU_COLS + BETA_COLS
ALPHA     = 0.05
sns.set_theme(style="whitegrid", font_scale=1.1)


# ══════════════════════════════════════════════════════════════════════════════
# 1.  CARGA Y PREPARACIÓN DE DATOS
# ══════════════════════════════════════════════════════════════════════════════

def cargar_datos(ruta_csv: str = "resultados_completos.csv") -> pd.DataFrame:
    """
    Carga el CSV y añade:
      - 'Tipo'  : Real / Imaginado / Reposo
      - 'Sujeto': alias limpio del campo 'Registro' (ej. 'sub_1' → 'S1')

    Parámetros
    ----------
    ruta_csv : str — ruta al CSV generado por el pipeline EEG.

    Retorna
    -------
    pd.DataFrame con columnas auxiliares añadidas.
    """
    df = pd.read_csv(ruta_csv)

    def clasificar_tipo(tarea: str) -> str:
        if "Real"      in str(tarea): return "Real"
        if "Imaginado" in str(tarea): return "Imaginado"
        return "Reposo"

    df["Tipo"]   = df["Tarea"].apply(clasificar_tipo)

    # Crear etiqueta de sujeto corta (S1, S2, …) para ejes más limpios
    sujetos_unicos = sorted(df["Registro"].unique(),
                            key=lambda s: int(s.split("_")[1]))
    mapa_sujeto = {s: f"S{i+1}" for i, s in enumerate(sujetos_unicos)}
    df["Sujeto"] = df["Registro"].map(mapa_sujeto)

    return df, sujetos_unicos, mapa_sujeto


# ══════════════════════════════════════════════════════════════════════════════
# 2.  ÍNDICE DE SUPRESIÓN POR SUJETO
# ══════════════════════════════════════════════════════════════════════════════

def calcular_indice_supresion(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calcula el Índice de Supresión (IS) para cada sujeto y canal:

        IS = (media_Reposo − media_Imaginado) / media_Reposo × 100

    Interpretación:
        IS > 0  →  la imaginería REDUCE la potencia respecto al reposo
                   (supresión esperada, indicador de buena imaginería motora)
        IS ≤ 0  →  sin supresión o incluso aumento de potencia

    Parámetros
    ----------
    df : DataFrame con columnas Tipo, Sujeto y columnas de potencia.

    Retorna
    -------
    pd.DataFrame con filas=sujeto, columnas=canales, valores=IS (%).
    """
    reposo    = df[df["Tipo"] == "Reposo"].groupby("Sujeto")[ALL_COLS].mean()
    imaginado = df[df["Tipo"] == "Imaginado"].groupby("Sujeto")[ALL_COLS].mean()

    # IS en porcentaje
    is_df = ((reposo - imaginado) / reposo * 100).round(2)
    is_df.index.name = "Sujeto"
    return is_df


# ══════════════════════════════════════════════════════════════════════════════
# 3.  ANÁLISIS ESTADÍSTICO
# ══════════════════════════════════════════════════════════════════════════════

def analizar_hipotesis_3(df: pd.DataFrame) -> dict:
    """
    Prueba estadística de la Hipótesis 3 para canales Mu y Beta:

        1. ANOVA de un factor (Sujeto) sobre datos de imaginería motora
        2. Post-hoc de Tukey si el ANOVA es significativo
        3. Eta² como tamaño de efecto

    Se analiza separadamente el subconjunto de tareas Imaginado.

    Parámetros
    ----------
    df : DataFrame con columna 'Sujeto' y columnas de potencia.

    Retorna
    -------
    dict con resultados por canal.
    """
    imagineria = df[df["Tipo"] == "Imaginado"].copy()
    sujetos    = sorted(imagineria["Sujeto"].unique())
    resultados = {}

    print("\n" + "═" * 65)
    print("  HIPÓTESIS 3 — Variabilidad inter-sujeto en imaginería motora")
    print("═" * 65)

    for canal in ALL_COLS:
        # ── Extraer grupos por sujeto ─────────────────────────────────
        grupos_data = [imagineria[imagineria["Sujeto"] == s][canal].dropna()
                       for s in sujetos]
        # Filtrar grupos con al menos 1 dato
        sujetos_val  = [s for s, g in zip(sujetos, grupos_data) if len(g) >= 1]
        grupos_val   = [g for g in grupos_data if len(g) >= 1]

        if len(grupos_val) < 2:
            print(f"\n  Canal {canal}: datos insuficientes.")
            continue

        n_total = sum(len(g) for g in grupos_val)
        k       = len(grupos_val)

        # ── ANOVA ─────────────────────────────────────────────────────
        F, p = stats.f_oneway(*grupos_val)
        ss_b = F * (k - 1)
        eta2 = ss_b / (ss_b + (n_total - k))
        sig  = "✅" if p < ALPHA else "❌"

        print(f"\n  Canal: {canal}")
        print(f"    ANOVA → F = {F:.3f} | p = {p:.4f} | η² = {eta2:.3f}  {sig}")

        # ── Post-hoc Tukey ────────────────────────────────────────────
        tukey_result = None
        if p < ALPHA:
            valores   = imagineria.loc[
                imagineria["Sujeto"].isin(sujetos_val), canal
            ].dropna()
            etiquetas = imagineria.loc[valores.index, "Sujeto"]
            tukey_result = pairwise_tukeyhsd(valores, etiquetas, alpha=ALPHA)
            # Solo mostrar pares significativos
            df_tukey = pd.DataFrame(
                data    = tukey_result._results_table.data[1:],
                columns = tukey_result._results_table.data[0]
            )
            sig_pares = df_tukey[df_tukey["reject"] == True]  # noqa: E712
            if len(sig_pares) > 0:
                print(f"    Pares significativos (Tukey, α={ALPHA}):")
                print(sig_pares[["group1", "group2", "meandiff", "p-adj"]].to_string(index=False))
            else:
                print("    (Ningún par alcanza significancia post-hoc)")

        resultados[canal] = {
            "F": F, "p": p, "eta2": eta2,
            "significativo": p < ALPHA,
            "tukey": tukey_result,
            "grupos_val": grupos_val,
            "sujetos_val": sujetos_val
        }

    return resultados


# ══════════════════════════════════════════════════════════════════════════════
# 4.  VISUALIZACIONES
# ══════════════════════════════════════════════════════════════════════════════

def plot_boxplot_sujetos(df: pd.DataFrame, resultados: dict):
    """
    Boxplot de la potencia Mu por sujeto durante imaginería motora,
    uno por canal. Destaca con color si el ANOVA fue significativo.
    """
    imagineria = df[df["Tipo"] == "Imaginado"]
    orden_suj  = sorted(imagineria["Sujeto"].unique())
    paleta     = sns.color_palette("tab10", n_colors=len(orden_suj))

    fig, axes = plt.subplots(3, 2, figsize=(16, 14))
    axes      = axes.flatten()
    fig.suptitle(
        "H3 — Potencia relativa por Sujeto durante Imaginería Motora\n"
        "(Canales Mu y Beta)",
        fontsize=13, fontweight="bold"
    )

    for ax, canal in zip(axes, ALL_COLS):
        res   = resultados.get(canal, {})
        p_val = res.get("p", np.nan)
        sig   = res.get("significativo", False)

        sns.boxplot(
            data=imagineria, x="Sujeto", y=canal, order=orden_suj,
            palette=paleta, width=0.6, linewidth=1.2,
            flierprops=dict(marker="o", markersize=3, alpha=0.4),
            ax=ax
        )
        sns.stripplot(
            data=imagineria, x="Sujeto", y=canal, order=orden_suj,
            color="black", alpha=0.3, size=3, jitter=True, ax=ax
        )

        color_titulo = "#1a7a1a" if sig else "#cc3300"
        ax.set_title(
            f"{canal}  —  p = {p_val:.4f}  {'✅' if sig else '❌'}",
            fontsize=10, color=color_titulo, fontweight="bold"
        )
        ax.set_xlabel("Sujeto")
        ax.set_ylabel("Densidad de potencia relativa")
        ax.tick_params(axis="x", rotation=30)

    # Ocultar panel sobrante si ALL_COLS tiene número impar
    for extra_ax in axes[len(ALL_COLS):]:
        extra_ax.set_visible(False)

    plt.tight_layout()
    plt.savefig("H3_boxplot_sujetos.png", dpi=150, bbox_inches="tight")
    print("\n  [Guardado] H3_boxplot_sujetos.png")
    plt.show()


def plot_ranking_supresion(is_df: pd.DataFrame):
    """
    Gráfica de barras horizontales que muestra el Índice de Supresión
    promedio (sobre canales Mu) por sujeto. Ordena de mayor a menor
    supresión para visualizar el 'ranking de imaginería motora'.
    """
    # IS promedio sobre los tres canales Mu
    is_mu = is_df[MU_COLS].mean(axis=1).sort_values(ascending=True)

    colores = ["#70AD47" if v > 0 else "#FF4444" for v in is_mu.values]

    fig, ax = plt.subplots(figsize=(9, 6))
    bars = ax.barh(is_mu.index, is_mu.values, color=colores,
                   edgecolor="black", linewidth=0.6, height=0.6)

    # Anotar valor en cada barra
    for bar, val in zip(bars, is_mu.values):
        ax.text(
            val + (0.3 if val >= 0 else -0.3),
            bar.get_y() + bar.get_height() / 2,
            f"{val:.1f}%", va="center",
            ha="left" if val >= 0 else "right", fontsize=9
        )

    ax.axvline(0, color="black", linewidth=1.2, linestyle="--")
    ax.set_xlabel("Índice de supresión Mu promedio (%)\n"
                  "(Reposo − Imaginado) / Reposo × 100", fontsize=10)
    ax.set_title(
        "H3 — Ranking de supresión Mu durante Imaginería Motora por Sujeto\n"
        "(Verde = supresión; Rojo = sin supresión / aumento)",
        fontsize=11, fontweight="bold"
    )

    verde  = mpatches.Patch(color="#70AD47", label="Supresión (IS > 0)")
    rojo   = mpatches.Patch(color="#FF4444", label="Sin supresión (IS ≤ 0)")
    ax.legend(handles=[verde, rojo], loc="lower right")

    plt.tight_layout()
    plt.savefig("H3_ranking_supresion.png", dpi=150, bbox_inches="tight")
    print("  [Guardado] H3_ranking_supresion.png")
    plt.show()





# ══════════════════════════════════════════════════════════════════════════════
# 5.  PUNTO DE ENTRADA
# ══════════════════════════════════════════════════════════════════════════════

if __name__ == "__main__":

    # 1. Cargar datos
    df, sujetos_unicos, mapa_sujeto = cargar_datos("resultados_completos.csv")
    print(f"  Dataset cargado: {df.shape[0]} filas × {df.shape[1]} columnas")
    print(f"  Sujetos detectados: {sujetos_unicos}")

    # 2. Calcular índice de supresión por sujeto
    is_df = calcular_indice_supresion(df)
    print("\n  Índice de supresión Mu (%) por sujeto:")
    print(is_df[MU_COLS].to_string())

    # 3. Análisis estadístico
    resultados = analizar_hipotesis_3(df)

    # 4. Visualizaciones
    print("\n  Generando gráficas...")
    plot_boxplot_sujetos(df, resultados)
    plot_ranking_supresion(is_df)

    print("\n  ✅ Análisis H3 completado.\n")
```
